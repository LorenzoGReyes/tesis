```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib
from math import sqrt
```

```python
r = np.arange(0.78,1.13,0.002)
np.random.shuffle(r)
delta = np.random.uniform(0,1, size=(175,))
theta = np.arange(0,2*np.pi,0.036)

x_1 = (r)*np.cos(theta+delta)
y_1 = (r)*np.sin(theta+delta)
data_1 = {'x_1':x_1, 'y_1':y_1}
df_1 = pd.DataFrame(data_1)
plt.figure(figsize=(5,5))
plt.plot(x_1,y_1,'.', color='black')
plt.axis('off')
plt.savefig('point_cloud_circle.png')
plt.show()
```

```python
r = np.arange(0,.35,0.002)
np.random.shuffle(r)
delta = np.random.uniform(0,1, size=(175,))
theta = np.arange(0,2*np.pi,0.036)

x_2 = (r)*np.cos(theta+delta)
y_2 = (r)*np.sin(theta+delta)
data_2 = {'x_2':x_2, 'y_2':y_2}
df_2 = pd.DataFrame(data_2)
plt.figure(figsize=(5,5))
plt.plot(x_2,y_2,'.', color='black')
plt.axis('off')
plt.ylim(-1, 1)
plt.xlim(-1,1)
plt.savefig('point_cloud_other.png')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3).fit(df_1)
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=["x_1", "y_1"])
df_1['predicted'] = kmeans.labels_
colors = ["red", "green", "blue"]
df_1['color'] = df_1['predicted'].map(lambda p: colors[p])
ax = df_1.plot(    
    kind="scatter", 
    x="x_1", y="y_1",
    figsize=(10,8),
    c = df_1['color']
)
centroids.plot(
    kind="scatter", 
    x="x_1", y="y_1", 
    marker=".", c=["r", "g", "b"], s=550,
    ax=ax
)
plt.axis('off')
```

```python
np.random.seed(1)
x = np.random.normal(0, 1, 150)
y = np.random.normal(0, 1, 150)

x[0:51] = x[0:51] + 5
y[0:51] = y[0:51] + 5
x[51:101] = x[51:101] - 1
x[101:151] = x[101:151] + 5
y[101:151] = y[101:151] - 5

data = {'x':x, 'y':y}
df = pd.DataFrame(data)
```

```python
df_1 = df.copy()
kmeans_1 = KMeans(n_clusters=1, n_init=20, random_state=1).fit(df_1)
centroids_1 = pd.DataFrame(kmeans_1.cluster_centers_, columns=["x", "y"])
df_1['predicted'] = kmeans_1.labels_
colors_1 = ["crimson"]
df_1['color'] = df_1['predicted'].map(lambda p: colors_1[p])
ax = df_1.plot(    
    kind="scatter", 
    x="x", y="y",
    figsize=(10,8),
    c = df_1['color'],
    s=50
)
centroids_1.plot(
    kind="scatter", 
    x="x", y="y", 
    marker="^", c=["crimson"], s=500,
    ax=ax
)
plt.axis('off')
plt.savefig('kmeadias1.png')
plt.show()
```

```python
range(1, 20)
k_means = [KMeans(n_clusters=i, n_init=20, random_state=1) for i in range(1,20)]
variance = [k_means[i].fit(df).inertia_ for i in range(len(k_means))]
```

```python
sns.set(style='darkgrid',)
ax = sns.lineplot(x=range(1,20), y=variance)
ax.set(xlabel='Valor para k', ylabel='Varianza')
ax.figure.set_size_inches(10,4)
ax.set_xticks(np.arange(0,21,1))
ax.figure.set_size_inches(16,6)
plt.savefig('codo.png')
plt.show()
```

```python
df2 = df.copy()
colores = ['crimson', 'gold', 'blue']
np.random.seed(seed=8)
lista = np.random.randint(0,3,150)
colores2 =[]


for i in range(0, 150):
    colores2.append(colores[lista[i]])


df2['colores'] = colores2
```

```python
ax = df2.plot(    
    kind="scatter", 
    x="x", y="y",
    figsize=(10,8),
    c = df2['colores'],
    s=50
)

plt.axis('off')
plt.savefig('nubededatos2.png')
plt.show()
```

```python
azules = df2['colores'] =='blue'
rojos = df2['colores'] == 'crimson'
amarillos = df2['colores'] == 'gold'
centroides = []

centroides.append((df2['x'][azules].mean(axis=0), df2['y'][azules].mean(axis=0)))
centroides.append((df2['x'][rojos].mean(axis=0), df2['y'][rojos].mean(axis=0)))
centroides.append((df2['x'][amarillos].mean(axis=0), df2['y'][amarillos].mean(axis=0)))

centroides = pd.DataFrame(centroides, columns=['x', 'y'])
```

```python
ax = df2.plot(    
    kind="scatter", 
    x="x", y="y",
    figsize=(10,8),
    c = df2['colores'],
    s=50
)
centroides.plot(
    kind="scatter", 
    x="x", y="y", 
    marker="^", c=["blue", "crimson", "gold"],
    s=750,
    ax=ax
)
plt.axis('off')
plt.savefig('nubededatos3.png')
plt.show()
```

```python
colores_nuevos = []

for i in range(0, 150):
    dist_b = sqrt(((df2['x'][i] - centroides['x'][0])**2) + ((df['y'][i] - centroides['y'][0])**2))
    dist_r = sqrt(((df2['x'][i] - centroides['x'][1])**2) + ((df['y'][i] - centroides['y'][1])**2))
    dist_y = sqrt(((df2['x'][i] - centroides['x'][2])**2) + ((df['y'][i] - centroides['y'][2])**2))
    if dist_b <= dist_r and dist_b <= dist_y:
        colores_nuevos.append('blue')
    elif dist_r <= dist_b and dist_r <= dist_y:
        colores_nuevos.append('crimson')
    else:
        colores_nuevos.append('gold')


df2['colores_nuevos'] = colores_nuevos
```

```python
ax = df2.plot(    
    kind="scatter", 
    x="x", y="y",
    figsize=(10,8),
    c = df2['colores_nuevos'],
    s=50
)
centroides.plot(
    kind="scatter", 
    x="x", y="y", 
    marker="^", c=["blue", "crimson", "gold"],
    s=750,
    ax=ax
)
plt.axis('off')
plt.savefig('nubededatos4.png')
plt.show()
```
